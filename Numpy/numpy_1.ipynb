```python
import matplotlib.pyplot as plt
import scipy.stats as sp
```

```python
def plot_col_desity(data, xlim = None, *args, **kwargs):
    desity_per_col = [sp.kde.gaussian_kde(col) for col in data.T]
    if xlim is not None:
        m, M = xlim
    else:
        m, M = data.min(), data.max()
    x = np.linspace(m, M, 100)

    fig, ax = plt.subplots()
    for density in desity_per_col:
        ax.plot(x, density(x), *args, **kwargs)
    ax.set_xlabel('log-counts')
    ax.set_ylabel('frequency')

    if xlim is not None:
        ax.set_xlim(xlim)
    
    plt.show()
```

```python
plot_col_desity(np.log(lrr +1))
```

```python
plot_col_desity(np.log(lrr_size +1), xlim=(0, 10))
```

```python
def sliding_window_1d(arr, window):
    result = []
    for i in range(len(arr) - window + 1):
        w = arr[i:i+window]
        result.append(w)
    return np.array(result)
```

```python
sliding_window_1d(np.array([1,2,3,4,5,6,7,8]), 3)
```
