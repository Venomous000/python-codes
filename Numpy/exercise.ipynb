2.14 Compute softmax stable: implement numerically stable softmax for logits.

```python
# simple softmax
logits = np.array([0.0000000000000000000000000000000000000001, 1.0, 0.1])
exp_logits = np.exp(logits)
softmax = exp_logits / np.sum(exp_logits)
print(softmax)
```

```python
# stable softmax
logits = np.array([0.0000000000000000000000000000000000000001, 1.0, 0.1])
m = np.max(logits)
exp_logits = np.exp(logits-m)
softmax = exp_logits / exp_logits.sum()
print(softmax)
```
