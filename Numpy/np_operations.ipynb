axis=(2,0)	    Same as (0,2)	                (3,)

#### Function Signature
numpy.dot(a, b, out=None)

dot behaves as:

üîπ 1D ¬∑ 1D ‚Üí scalar (inner product)
dot(ùëé,ùëè)=‚àëùëéùëñùëèùëñ

üîπ 2D ¬∑ 2D ‚Üí matrix multiplication
ùê∂ùëñùëó=‚àëùëò (ùê¥ùëñùëò)(ùêµùëóùëò)  

üîπ 1D ¬∑ 2D ‚Üí row vector * matrix


Treat vector as shape (1, n)

üîπ 2D ¬∑ 1D ‚Üí matrix * column vector

Treat vector as shape (n, 1)

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.dot(a, b))
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[10, 20], [30, 40]])

print(np.dot(A, B))
```

```python
v = np.array([1, 2])
M = np.array([[3, 4],
              [5, 6]])

print(np.dot(v, M))
print(np.dot(M, v))
```

dot() behavior

(n,) ¬∑ (n,)	inner product

(n,) ¬∑ (n,m)	row √ó matrix

(n,m) ¬∑ (m,)	matrix √ó column

(n,m) ¬∑ (m,k)	matrix multiplication

(..., m) ¬∑ (m, ...)	generalized contraction

### Function Signature
numpy.matmul(x1, x2, /, out=None, *, casting='same_kind', order='K', dtype=None,subok=True, signature=None, extobj=None)

A.shape[-1] == B.shape[-2]

The last dimension of A must equal the second-to-last of B.

Examples:

(m, n) @ (n, p) ‚Üí (m, p)

(3,) @ (3,) ‚Üí dot product, result: scalar

(3,) @ (3,5) ‚Üí result: (5,) ‚Üê vector √ó matrix

(4,3) @ (3,) ‚Üí result: (4,) ‚Üê matrix √ó vector



```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])   # shape (2, 3)

B = np.array([[10, 11],
              [20, 21],
              [30, 31]])    # shape (3, 2)

C = A @ B

print(C)
print(C.shape)
```

Softmax

Softmax is used in ML to convert logits into probabilities:

```python
logits = np.array([0.0000000000000000000000000000000000000001, 1.0, 0.1])
exp_logits = np.exp(logits)
softmax = exp_logits / np.sum(exp_logits)
print(softmax)
```

### Function Signature
numpy.allclose(a, b, rtol=1e-05, atol=1e-08, equal_nan=False)

Parameters:

a, b ‚Üí input arrays to compare.

rtol ‚Üí relative tolerance: abs(a-b) <= (atol + rtol * abs(b)).

atol ‚Üí absolute tolerance.

equal_nan ‚Üí if True, NaNs in the same position are considered equal.

Returns: True or False.

```python
a = np.array([1.0, 2.0, 3.0])
b = np.array([1.0, 2.000001, 2.999999])
print(np.allclose(a, b))
```

```python
a = np.array([1.0, np.nan])
b = np.array([1.0, np.nan])

print(np.allclose(a, b))             
print(np.allclose(a, b, equal_nan=True)) 
np.allclose([1e-9], [0], rtol=0, atol=1e-8)
```
