#### MultiIndex / hierarchical indexing

```python
arrays = [
    ["A", "A", "B", "B", "C"],
    [1, 2, 3, 1, 2]
]

multi_index = pd.MultiIndex.from_arrays(arrays, names=("Group", "Number"))

df_multi = pd.DataFrame({"Value": [10, 20, 30, 40, 50]}, index=multi_index)
df_multi
```
